```python
import json
import os

json_name = 'Image_label.json'

with open(json_name,"r") as f:
    load_prompt = json.load(f)
    
lst_image = os.listdir('Image')
num_image = 0
for i in range(len(load_prompt)):
    for j in range(len(lst_image)):
        if lst_image[j] == load_prompt[i]['img']:
            num_image = num_image + 1
print(num_image)
print(len(load_prompt))
print(len(lst_image))
```

```python
import copy

label_txt = '单人/双人/小团体（3-5人）/大团体（6-10人）/大群体（10人以上）/朋友/情侣/闺蜜/亲子/商业伙伴/宠物/暖色调/冷色调/饱和度高/饱和度低/明度亮/明度暗/春节/万圣节/圣诞节/情人节/端午节/中秋节/婚礼/毕业/生日/旅行/汉服/波西米亚/JK/和风/美式街头/西装/休闲装/cosplay/森系/家庭/影棚/咖啡馆/图书馆/精品店/教室/CBD/历史街区/天台/地铁站/桥梁/街道/中式园林/公园草地/湖泊/路牌/动物园/迪士尼/漫展/自然/清晨/正午/黄昏/夜晚/黄金时刻/晴天/雾霾/阴天/绵绵小雨/大雨/飘雪/大雪/极光/花束/乐器/书籍/泡泡机/粉笔/气球/玩偶/笔'
label_lst = label_txt.split('/')
label_dic = {}
for i in range(len(label_lst)):
    label_dic[label_lst[i]] = None
print(label_dic)

change_label = []
for i in range(len(load_prompt)):
    label_dict = copy.copy(label_dic)
    img = load_prompt[i]['img']
    label = load_prompt[i]['label']
    label_item = []
    for key in label:
        if label[key] != None:
            for item in label[key]:
                label_item.append(item)
    print(label_item)
    for key_new in label_item:
        if key_new in label_dict:
            label_dict[key_new] = True
    change_label.append({'img':img,'label':label_dict})
    print(label_dict)
```

```python
json_name = 'Image_label_change.json'
with open(json_name,"w") as f:
    json.dump(change_label,f)
```

```python
import json

def get_image_label(label,json_name):
    with open(json_name,"r") as f:
        load_prompt = json.load(f)
    lst_image = []
    for item in label:
        #print(item)
        for j in range(len(load_prompt)):
            img = load_prompt[j]['img']
            label_dic = load_prompt[j]['label']
            if item in label_dic:
                if label_dic[item] == True:
                    lst_image.append(img)
                    #print(label_dic[item])
    return lst_image

label = ['单人']
json_name = 'Image_label_change.json'
lst_image = get_image_label(label,json_name)
print(lst_image)
```
